# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import models, layers
from keras.utils.vis_utils import model_to_dot
from keras.utils import plot_model
import keras

from sklearn.model_selection import train_test_split
```

### Initialising Global Variables

```python
IMAGE_SIZE = 32
BATCH_SIZE = 128
EPOCH = 15
```

# EDA on Devanagiri Character Dataset

The Dataset consists of 92000 rows. Each columns represents the pixel information for the 32x32 image. We have to predict the label for each image which is stored under the column 'character'.

The size of the dataset was around 250mb. As all the data represents pixel information, the values in each column ranges from 0 to 255. I decided to downcast the values to unsingned dtypes - **uint8**. The memory usage reduced to just **90.5 mb**.

Reading in the numeric data first with dtype = 'uint8'. Then we will get the character column. We could have done this in one single line but the dtype option in pandas takes in dictionaries but in the format {column_1: dtype, columns_2:dtype, ....}. I did not wanted to create a dictionary without having the columns names first and then using a roundabout to create one.

```python
df = pd.read_csv("data.csv", usecols=lambda x: x != 'character', dtype = np.uint8)
df['character'] = pd.read_csv("data.csv", usecols=['character'], dtype = 'category')
df.head()
```

```python
df.info()
```

Next we check for any NA values.

```python
np.sum(df.isna().sum()>0)
```

Since, we don't have any missing or na values, we can now look at our data.

```python
characters = df.character.unique()
characters
```

```python
# Remove the extra words from the target word. We'll keep only the last part and remove the prefix.

df['character'] = df.character.str.rsplit('_', 1, expand=True).iloc[:, -1]
```

Let's check whether the data is balanced or not.

```python
df.character.value_counts()
```

```python
class_names = df.character.unique()
```

We can see that the data is balanced. Now  we can move further onto our analysis but first let's take a look at our data.

```python
X_images = df.iloc[:, :-1].values.reshape((-1,32,32))
fig, ax = plt.subplots(1,4, figsize = (12,9))

for i, j in enumerate(np.random.randint(92000, size=(4))):
    ax[i].imshow(X_images[j])
    ax[i].title.set_text(df.iloc[j, -1])

plt.show()
```

# Preprocessing

```python
X, y = df.drop('character', axis = 1), df.character
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y)
```

Since our y contains categorical data. We'll transform it in order to feed into the model. We've used LabelBinarizer() here because it is easier to transform then untransform after the model has predicted in binary label format to the single multilabel format. We could have also used OneHotEncoder.

```python
from sklearn.preprocessing import LabelBinarizer
labelbinarize = LabelBinarizer()
train_y = labelbinarize.fit_transform(train_y)
```

```python
train_X = np.divide(train_X, 255)
test_X = np.divide(test_X, 255)
```

```python
train_X = train_X.values.reshape(-1, 32, 32, 1)
test_X = test_X.values.reshape(-1, 32, 32, 1)
```

```python
print(X.shape)
print(train_X.shape)
print(test_X.shape)
print(train_y.shape)
print(test_y.shape)
```

# Model

## Model Building

```python
num_classes = train_y.shape[1]
input_shape=(IMAGE_SIZE, IMAGE_SIZE, 1)

model = tf.keras.Sequential([
        
  layers.Conv2D(128, (4,4), activation='relu', input_shape=input_shape),
  layers.MaxPooling2D((2,2)),
  layers.Dropout(0.2),
        
  layers.Conv2D(32,(3,3), activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Dropout(0.2),
        
  layers.Conv2D(32, (2,2), activation='relu'),
  layers.MaxPooling2D((2,2)),
  layers.Dropout(0.2),
        
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes, activation='softmax'),
])

model.build(input_shape=input_shape)
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
# Use Early Stopping to save the model from Overfitting and stop executing when no progress is made
# We'll use checkpoints to save the best model in out training.

from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    min_delta=0.05, 
    patience = 2, 
    restore_best_weights=True,
)

checkpoint_path = "weights.best.hdf5"
checkpoint = keras.callbacks.ModelCheckpoint(checkpoint_path,
                             monitor='val_accuracy',
                             verbose=1,
                             save_best_only=True,
                             mode='max')
```

## Training the Model

```python
history = model.fit(
    train_X,
    train_y,
    epochs = EPOCH,
    batch_size = BATCH_SIZE,
    verbose = 1,
    validation_split=0.2,
    callbacks = [early_stopping, checkpoint]
)
```

```python
history.history.keys()
```

```python
from operator import pos
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))
```

```python
plt.figure(
  figsize=(12,4)
)
plt.subplot(1,2,1)
plt.plot(epochs, acc, label = 'Training Accuracy')
plt.plot(epochs, val_acc, label = 'Validation Accuracy')
plt.legend(loc = 'lower right')
_ = plt.title("Training Vs Validation Accuracy")
```

```python
plt.figure(
  figsize=(12,4)
)
plt.subplot(1,2,1)
plt.plot(epochs, loss, label = 'Training Loss')
plt.plot(epochs, val_loss, label = 'Validation Loss')
plt.legend(loc = 'upper right')
_ = plt.title("Training Vs Validation Loss")
```

## Model Evaluation

Let's Load the best model so we can use it for prediction

```python
model = keras.models.load_model('weights.best.hdf5')
```

```python
predictions = model.predict(test_X)
predictions.shape
```

```python
preds = labelbinarize.inverse_transform(predictions)
```

```python
preds.shape
```

Let's take a look at our predictions.

```python
X_images = test_X.reshape((-1,32,32))
fig, ax = plt.subplots(1,4, figsize = (12,9))

for i, j in enumerate(np.random.randint(23000, size=(4))):
    ax[i].imshow(X_images[j])
    string = 'Actual Label: {0}\nPredicted Label: {1}'.format(test_y.iloc[j], preds[j])
    ax[i].title.set_text(string)
    
plt.show()
```

```python
from sklearn.metrics import classification_report, f1_score, confusion_matrix
```

```python
print(classification_report(test_y, preds))
```

```python
f1_score(test_y, preds, average = 'weighted')
```

```python
fig = plt.figure(figsize = (14, 14))
matrix = confusion_matrix(test_y, preds)
# Create pandas dataframe
dataframe = pd.DataFrame(matrix, index=class_names, columns=class_names)
# Create heatmap
sns.heatmap(dataframe, annot=True, cbar=None, cmap="Blues")
plt.title("Confusion Matrix"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```
